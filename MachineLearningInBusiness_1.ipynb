STEP 4

```python
# Writing a function to calculate profit from top oil producing wells from a set of 500 randomly selected oil wells that were obtained in model predictions
import numpy as np

def calculate_profit(predicted_volumes,target_values,wells_count,budget,revenue_per_unit_product, sample_size=None):
    # Sampling logic
    if sample_size:
        sample_indices=np.random.choice(len(predicted_volumes), size=sample_size, replace=False)
        predicted_volumes=predicted_volumes[sample_indices]
        target_values= target_values.iloc[sample_indices]
        target_values= target_values.reset_index(drop=True)
        
    predictions_valid_series=pd.Series(predicted_volumes)
    top_wells_predictions=predictions_valid_series.nlargest(wells_count)
    target_values_series=pd.Series(target_values, index=predictions_valid_series.index)
    top_target_wells=target_values_series[top_wells_predictions.index]
    sum_top_wells_target=top_target_wells.sum()

    profit=(sum_top_wells_target*revenue_per_unit_product)-budget
    return profit

# Calculating Profit for all three Regions
profit_0=calculate_profit(predictions_valid_0,target_valid_0,200,100000000,4500, sample_size=500)
profit_1=calculate_profit(predictions_valid_1,target_valid_1,200,100000000,4500, sample_size=500)
profit_2=calculate_profit(predictions_valid_2,target_valid_2,200,100000000,4500, sample_size=500)

print(f"Predicted profits from top 200 oil producing wells in Regions 0,1 and 2 is:\n{profit_0:.2f}, {profit_1:.2f}, {profit_2:.2f} respectively")


```

Analysis so far:

When we select top 200 wells - from 500 randomly chosen wells from each region, we get predicted profits as: $6.45 million for Region 0, $6.56 million for Region 1 and $2.84 million for region 2. This suggests that Region 1 has highest profits, followed by Region 0, with Region 2 having the lowest profit. All regions are profitable(positive values).
To get more confidence about our calculations, we will now use bootstrapping which will utilize sampling multiple times to get a distribution of profit.


STEP 5

```python
# Using bootstrapping technique with 1000 samples to find distribution of profit

import numpy as np

state=np.random.RandomState(12345)

# Applying Bootstrapping Technique by defining a function

def bootstrapping(predictions_valid,target_valid):
    revenue_values=[]
    # Creating a validation dataset with target vaues of predicted set
    validation_data=pd.DataFrame({'predictions': predictions_valid, 'actual': target_valid}).reset_index(drop=True)

    # Creating a loop for creation of 1000 bootstrap samples
    for i in range(1000):
        # Selecting 500 random points for a sample
        validation_data_subsample=validation_data.sample(n=500,replace=True,random_state=i)
        #Extracting prediction and target subsamples from each bootstrap sample
        predictions_subsample=validation_data_subsample['predictions']
        target_subsample=validation_data_subsample['actual']
        # Calculating revenue by selecting best 200 points for profit calculation
        revenue_values.append(calculate_profit(predictions_subsample,target_subsample,200,100000000,4500))
    
    revenue_values= pd.Series(revenue_values)
    mean=revenue_values.mean()
    # Calculating lower and upper quantiles for risk assesment
    lower_quantile_region=revenue_values.quantile(q=0.025)
    higher_quantile_region=revenue_values.quantile(q=0.975)

    # Calculating risk of losses
    risk_of_loss=(revenue_values<0).mean()
    
    print(f"Mean profit: {mean:.2f}")
    print(f"Lower quantile of confidence interval: {lower_quantile_region:.2f}")
    print(f"Higher quantile of confidence interval:{higher_quantile_region:.2f}")
    print(f"Risk of losses: {risk_of_loss:.1%}\n")



# Displaying all Bootstrapping Analysis with 95% confidence interval values:

print("-"*50)
print("Region 0 Bootstrapping Analysis:")
print("-"*50)
bootstrapping(predictions_valid_0, target_valid_0)

print("-"*50)
print("Region 1 Bootstrapping Analysis:")
print("-"*50)
bootstrapping(predictions_valid_1, target_valid_1)

print("-"*50)
print("Region 2 Bootstrapping Analysis:")
print("-"*50)
bootstrapping(predictions_valid_2, target_valid_2)

```
